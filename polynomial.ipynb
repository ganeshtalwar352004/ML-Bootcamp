```python
def power_function(n,X):
   
    df = pd.DataFrame()
    for n in range(1,n+1):
        for i in range(n+1):
            for j in range(n+1-i):
                x=[]
                x=(X[:,0]**i)*(X[:,1]**j)*(X[:,2]**(n-i-j))
#             X = np.append(X, x, axis=1)
                l=[]
                l=np.array(x)
                l=l.T
                x11=l.tolist()
                df[f'a^{i}*b^{j}*c^{n-i-j}']=x11
    return df
    
```

```python
def output_label(x,w,b):
    label=[]
    m,n=x.shape
    for i in range(m):
        f_x=(np.dot(w,x[i]))+b
        label.insert(i,f_x)
    pdf['label1']=label
    return pdf
```

```python
X=output_label(X_train_2,q2,s2)
X1=X[['label','label1']]
print(X1.head(5))
x11=X1.to_numpy()
print(x11.shape)
```

```python
def calculate_r2(actual_target_values,predicted_target_values):
    sum_squared_residuals=np.sum(np.square(predicted_target_values-actual_target_values))
    sum_squares=np.sum(np.square(np.mean(actual_target_values)-actual_target_values))
    r2=1-(sum_squared_residuals/sum_squares)
    return r2
```

```python
r2_score=calculate_r2(x11[:,0],x11[:,1])
print(r2_score)
```
