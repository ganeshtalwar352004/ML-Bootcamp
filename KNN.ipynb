```python
import numpy as np
import pandas as pd
from mlal1 import *
import math
```

```python
pdf=pd.read_csv('classification_train.csv')
arr=pdf.to_numpy()
X_train=arr[0:10000,2:786]
y_train=arr[0:10000,1]
x=arr[0:10000,1:786]
print(x[0:10000,1:785])
```

```python
pdf2=pd.read_csv('classification_test.csv')
arr2=pdf2.to_numpy()
X_test=arr2[0:10000,1:785]
id_test=arr2[0:10000,0]
```

```python
print(X_train.shape)
print(X_test.shape)
print(id_test.shape)
```

```python
def cal_eucleadian_distance(x1,x2):
    return np.sqrt(np.sum(np.square(x1-x2),axis=1))
```

```python
def cal_pred(x1,x2):
    k=100
    distance=cal_eucleadian_distance(x1[:,1:785],x2)
    indices=np.argsort(distance)
    predicted_value=np.mean(x1[indices[0:k],0])
    return predicted_value
```

```python
predictions=[]
m,n=X_test.shape
for i in range(m):
    predicted_value=cal_pred(x,X_test[i,:])
    
    pred=math.modf(predicted_value)
    if(pred[0]>=0.5):
        predictions.append(pred[1]+1)
    else:
        predictions.append(pred[1])
print(predictions)
pdf2['label1']=predictions
```
