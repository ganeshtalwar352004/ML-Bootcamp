```python
def derv_relu(Z):
    return Z>0
```

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.W1 = 0.001*np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = 0.001*np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))
    
    def forward(self, X):
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = np.tanh(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = np.exp(self.z2) / np.sum(np.exp(self.z2), axis=1, keepdims=True)
        return self.a2
    
    def backward(self, X, Y, output):
        error_output = output - Y
        delta_output = error_output
        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * derv_relu(self.a1)
        self.W2 -= 0.0001 * np.dot(self.a1.T, delta_output)
        self.b2 -= 0.0001 * np.sum(delta_output, axis=0, keepdims=True)
        self.W1 -= 0.0001 * np.dot(X.T, delta_hidden)
        self.b1 -= 0.0001 * np.sum(delta_hidden, axis=0)
    
    def train(self, X, Y):
        output = self.forward(X)
        self.backward(X, Y, output)
    
    def predict(self, X):
        output = self.forward(X)
        return np.argmax(output, axis=1)

# Create the neural network object and train it on the MNIST dataset
nn = NeuralNetwork(784, 100, 10)

for epoch in range(15):
    print('Epoch', epoch+1)
    for i in range(len(train_features)):
        nn.train(train_features[i:i+1], train_labels[i:i+1])
    predictions = nn.predict(train_features)
    accuracy = np.mean(predictions == np.argmax(train_labels, axis=1))
    print('Accuracy:', accuracy)

# Visualize some of the predictions
```
